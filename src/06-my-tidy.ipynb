### MELT
Melt is used to unpivot data. For instance, when I download data from Barclays, the countries are separated on a column by column (wide format) but to do proper analysis I need them on a normalized format (long)

```python
import pandas as pd
df_cape_wide = pd.read_csv('..\data\CAPE_Historical_Data_Barclays.csv') 
df_cape_wide.head()
```

```python
df_cape_long = pd.melt(df_cape_wide, id_vars=['Date'], var_name='Ticker', value_name='CAPE')
df_cape_long.head()
```
